```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../../datasets/supervised/diabetes_clean.csv")
```

```python
df
```

```python
# Separar as características e a variável de resposta
X = df.drop("diabetes", axis=1)
y = df["diabetes"]
```

```python
# Obter o número de características
num_features = X.shape[1]
```

```python
# Criar subplots
fig, axes = plt.subplots(nrows=num_features, figsize=(5, 5 * num_features))

# Loop sobre cada característica
for i, column in enumerate(X.columns):
    axes[i].scatter(X[column], y)
    axes[i].set_title(f'Scatter plot of {column} vs diabetes')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Diabetes')

# Ajustar o layout
plt.tight_layout()
plt.show()
```
