```python
# Endere√ßo para plotar legal.
# https://matplotlib.org/stable/plot_types/index.html
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# plt.style.use('Solarize_Light2')
x = np.linspace(0, 5, 100)
y = x ** 2
plt.plot(x, y)
plt.show()
```

```python
# V√°rias configura√ß√µes b√°sicas.
plt.plot(x, y, 'r') # Mudar a cor da parada.
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('T√≠tulo Isano')
plt.show()
```

## Subplots

```python
# Subplots
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(x, y, 'g')

plt.show()
```

```python
fig = plt.figure()

# Os dois primeiros n√∫meros s√£o referentes ao eixo a partir do qual tudo come√ßa.
# Os dois n√∫meros subsequentes s√£o a escala de crescimento.
axis1 = fig.add_axes([0,0,0.5,0.5])     # Come√ßa em x=0.0, y=0.0, e cresce numa escala de 0.5 para ambas as dire√ß√µes.
axis2 = fig.add_axes([0.6,0,0.5,0.5])   # Come√ßa em x=0.6, y=0.0, e cresce numa escala de 0.5 para ambas as dire√ß√µes.

axis1.plot(x, y/2 + 2)
axis2.plot(x, y*np.sin(2*np.pi))
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,2))

axes[0].plot(x, y)
axes[0].set_title('First Plot', color="red")
axes[1].plot(2*y, x)
axes[1].set_title('Second Plot', color="red")
plt.tight_layout()
```

## Logaritmo

```python
import matplotlib.pyplot as plt

x = list(range(0,1000))
y = [ element ** 2 for element in x]

plt.plot(x, y)
plt.show()
```

```python
import numpy as np

x = np.arange(0, 1000)
y = x ** 2

plt.plot(x, y)
plt.show()

plt.xscale('log')
plt.plot(x, y)
plt.show()
```
