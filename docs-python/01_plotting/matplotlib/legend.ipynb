```python
# Vamos importar a função randn que cria números aleatórios como um array com a dimensão que a gente quiser.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x = np.arange(0, 1000)
y = 10 * np.sin(x/100)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,2))

axes[0].plot(x, y)
axes[0].set_title('First Plot', color="red")
axes[1].plot(2*y, x)
axes[1].set_title('Second Plot', color="red")
plt.show()
```

```python
x = np.arange(0, 1000)
y = 10 * np.sin(x/100)

plt.plot(x, y,
    color='red',
    linestyle=':',
    label='meu labinho'
)
plt.legend()
plt.show()
```
