```python
import numpy as np
import matplotlib.pyplot as plt
```

# Templates
Vamos dar uma olhada nos estilos que temos disponíveis.

```python
plt.style.available
```

Vamos criar uma função e ver como as coisas funcionam.

```python
x = np.arange(0, 1000)
y = 10 *np.sin(x/100)
```

```python
plt.style.use('fivethirtyeight')
plt.plot(x, y)
```

```python
plt.style.use('ggplot')
plt.plot(x, y)
```

```python
plt.style.use('seaborn')
plt.plot(x, y)
```

```python
plt.style.use('default')
plt.plot(x, y)
```

# Estilo de linhas e Marcadores

```python
plt.plot(x, y,
    color="red",
    linestyle=":",
    label="legenda"    
)
```
